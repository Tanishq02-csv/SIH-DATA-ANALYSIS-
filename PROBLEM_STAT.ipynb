```python
#Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

ps_data=pd.read_csv('PS.csv')
#Renaming PS ID Column
ps_data=ps_data.rename(columns={"PS Number":"PS ID"})
#Removing serial no. coumn
ps_data=ps_data.drop("SERIAL No.",axis=1)
ps_data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute distributions
category_distribution = ps_data['Category'].value_counts()
theme_distribution = ps_data['Theme'].value_counts()

# Create subplots
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(14, 12))

# Custom color palettes
category_palette = sns.color_palette("coolwarm", len(category_distribution))
theme_palette = sns.color_palette("cubehelix", len(theme_distribution))

# Plot for Category
sns.countplot(
    data=ps_data, 
    y='Category', 
    ax=ax[0], 
    order=category_distribution.index, 
    palette=category_palette
)
ax[0].set_title('Distribution of Problem Statement Categories', fontsize=16, weight='bold', color='darkred')
ax[0].set_xlabel('Number of Problem Statements', fontsize=12, weight='bold')
ax[0].set_ylabel('Category', fontsize=12, weight='bold')

# Add value annotations for Category
for p in ax[0].patches:
    ax[0].annotate(
        f'{int(p.get_width())}', 
        (p.get_width() + 1, p.get_y() + p.get_height() / 2),
        ha='left', va='center', fontsize=10, color='black', weight='bold'
    )

# Plot for Theme
sns.countplot(
    data=ps_data, 
    y='Theme', 
    ax=ax[1], 
    order=theme_distribution.index, 
    palette=theme_palette
)
ax[1].set_title('Distribution of Problem Statement', fontsize=16, weight='bold', color='darkred')
ax[1].set_xlabel('Number of Problem Statements', fontsize=12, weight='bold')
ax[1].set_ylabel('Themes', fontsize=12, weight='bold')

# Add value annotations for Theme
for p in ax[1].patches:
    ax[1].annotate(
        f'{int(p.get_width())}', 
        (p.get_width() + 1, p.get_y() + p.get_height() / 2),
        ha='left', va='center', fontsize=10, color='black', weight='bold'
    )

# Add gridlines for better readability
for a in ax:
    a.grid(axis='x', linestyle='--', alpha=0.7)
    a.spines['top'].set_visible(False)
    a.spines['right'].set_visible(False)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set plot style
sns.set_theme(style="whitegrid")

# Create the boxplot
plt.figure(figsize=(14, 10))
palette = sns.color_palette("coolwarm", len(Theme_distribution))  # Gradient color palette

sns.boxplot(
    data=ps_data,
    y='Theme',
    x='Submitted Idea(s) Count',
    order=Theme_distribution.index,
    palette=palette
)

# Customizing the title and labels
plt.title('Distribution of Submitted Ideas by Theme', fontsize=16, weight='bold', color='darkblue')
plt.xlabel('Number of Submitted Ideas', fontsize=12, weight='bold')
plt.ylabel('Themes', fontsize=12, weight='bold')

# Add gridlines for better readability
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the histogram
plt.figure(figsize=(12, 6))
ax = sns.histplot(ps_data['Submitted Idea(s) Count'], bins=30, kde=True, color='skyblue', edgecolor='black')

# Add title and labels
plt.title('Distribution of Submitted Idea(s) Count', fontsize=16, weight='bold', color='darkblue')
plt.xlabel('Number of Submitted Ideas', fontsize=12, weight='bold')
plt.ylabel('Number of Problem Statements', fontsize=12, weight='bold')

# Annotate each bar with its height
for patch in ax.patches:
    height = patch.get_height()
    if height > 0:  # Only annotate if the height is greater than zero
        plt.text(
            patch.get_x() + patch.get_width() / 2,  # x-coordinate (center of the bar)
            height + 0.5,  # y-coordinate (slightly above the bar)
            f'{int(height)}',  # Text to display
            ha='center', va='center', fontsize=10, color='black', weight='bold'
        )

# Adjust layout for better appearance
plt.tight_layout()

# Show the plot
plt.show()
```
